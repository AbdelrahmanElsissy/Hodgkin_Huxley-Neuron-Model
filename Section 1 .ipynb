```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the equations and constants
Voltage_range = np.linspace(-100e-3, 100e-3, 1000)
Temperature_1 = 6.3
Temperature_2 = 28

k1 = 3**(0.1 * (Temperature_1 - 6.3))
k2 = 3**(0.1 * (Temperature_2 - 6.3))

# Alphas and betas
alpha_m = 1e3 * ((2.5 - 100 * Voltage_range) / (np.exp(2.5 - 100 * Voltage_range) - 1))
beta_m = 4e3 * np.exp(-500 * Voltage_range / 9)

alpha_n = 1000 * ((0.1 - 10 * Voltage_range) / (np.exp(1 - 100 * Voltage_range) - 1))
beta_n = 125 * np.exp(-25 * Voltage_range / 2)

alpha_h = 70 * np.exp(-50 * Voltage_range)
beta_h = 1000 / (np.exp(3 - 100 * Voltage_range) + 1)

# Time constants and steady-state values
tau_m1 = 1 / (k1 * (alpha_m + beta_m))
m_inf = alpha_m / (alpha_m + beta_m)
tau_m2 = 1 / (k2 * (alpha_m + beta_m))

tau_n1 = 1 / (k1 * (alpha_n + beta_n))
n_inf = alpha_n / (alpha_n + beta_n)
tau_n2 = 1 / (k2 * (alpha_n + beta_n))

tau_h1 = 1 / (k1 * (alpha_h + beta_h))
h_inf = alpha_h / (alpha_h + beta_h)
tau_h2 = 1 / (k2 * (alpha_h + beta_h))

Voltage_in_mV = Voltage_range * 1e3
```

```python
# Plots for T = T1 = 6.3ºC
tau_m_p1 = tau_m1 * 1e3
tau_n_p1 = tau_n1 * 1e3
tau_h_p1 = tau_h1 * 1e3

plt.figure()
plt.plot(Voltage_in_mV, tau_m_p1, label="$\\tau_m$")
plt.plot(Voltage_in_mV, tau_n_p1, label="$\\tau_n$")
plt.plot(Voltage_in_mV, tau_h_p1, label="$\\tau_h$")
plt.xlabel("Voltage (mV)")
plt.ylabel("Time Constant (ms)")
plt.legend()
plt.xlim([-100, 100])
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('Fig 1 graphtauT1.png', format='png', bbox_inches='tight')

plt.figure()
plt.plot(Voltage_in_mV, m_inf, label="$m_\\infty$")
plt.plot(Voltage_in_mV, n_inf, label="$n_\\infty$")
plt.plot(Voltage_in_mV, h_inf, label="$h_\\infty$")
plt.xlabel("Voltage (mV)")
plt.ylabel("Opening Probability")
plt.legend()
plt.xlim([-100, 100])
plt.ylim([0, 1])
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('Fig 3 graphmnh1.png', format='png', bbox_inches='tight')
```

```python
# Plots for T = T2 = 28ºC
tau_m_p2 = tau_m2 * 1e3
tau_n_p2 = tau_n2 * 1e3
tau_h_p2 = tau_h2 * 1e3

plt.figure()
plt.plot(Voltage_in_mV, tau_m_p2, label="$\\tau_m$")
plt.plot(Voltage_in_mV, tau_n_p2, label="$\\tau_n$")
plt.plot(Voltage_in_mV, tau_h_p2, label="$\\tau_h$")
plt.xlabel("Voltage (mV)")
plt.ylabel("Time Constant (ms)")
plt.legend()
plt.xlim([-100, 100])
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('Fig 2 graphtauT2.png', format='png', bbox_inches='tight')

plt.figure()
plt.plot(Voltage_in_mV, m_inf, label="$m_\\infty$")
plt.plot(Voltage_in_mV, n_inf, label="$n_\\infty$")
plt.plot(Voltage_in_mV, h_inf, label="$h_\\infty$")
plt.xlabel("Voltage (mV)")
plt.ylabel("Opening Probability")
plt.legend()
plt.xlim([-100, 100])
plt.ylim([0, 1])
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('Fig 4 graphmnh2.png', format='png', bbox_inches='tight')

plt.show()
```
